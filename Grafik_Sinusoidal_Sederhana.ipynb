```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# Data plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='Waktu (s)', ylabel='Voltase (mV)',
       title='Membuat grafik sederhana')
ax.grid()
plt.show()
```
